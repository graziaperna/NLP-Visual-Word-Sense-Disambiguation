```python
# import libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from cleantext import clean

# load messages dataset
data = pd.read_csv('Dataset\edos_labelled_aggregated.csv')
data.head()
```

```python
#df = pd.DataFrame(data)
#df=df.drop('split', axis = 1)
```

```python
df = pd.DataFrame(data)
df['cleaned_text'] = df.text.apply(clean,no_emoji=True)
```

```python
df['text'] = df.loc[df['text']!='', 'cleaned_text'].map(clean)
```

```python
df= df.replace(["not sexist"],0)
df= df.replace(["sexist"],1)
df=df.drop('cleaned_text', axis = 1)
df=df.drop('split', axis = 1)
df.head()
```

```python
try:
    df.insert(5, 'Threats', 0, allow_duplicates=False)
    df.insert(6, 'Derogation', 0, allow_duplicates=False)
    df.insert(7, 'Animosity', 0, allow_duplicates=False)
    df.insert(8, 'Prejudiced discussion', 0, allow_duplicates=False)
except:
    print("Columns already exist")
df.head()
```

```python
df.loc[df.label_category=="1. threats",'Threats']='1'
df.loc[df.label_category=="2. derogation",'Derogation']='1'
df.loc[df.label_category=="3. animosity",'Animosity']='1'
df.loc[df.label_category=="4. prejudiced discussion",'Prejudiced discussion']='1'

df.iloc[:10, :10] 
```
